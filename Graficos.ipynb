Modelos Graficos

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
array = np.arange(0, 10, 0.01)
plt.plot(array, array**2)
```

```python
array = np.arange(0, 10, 0.01)

plt.style.use('classic')

plt.plot(array, array**2)
```

```python
x = np.arange(0.5,10, 0.001)

y1 = np.log(x)
y2 = 5 * np.sin(x) / x 

#Escolhe o modelo do gráfico
plt.style.use('ggplot')

plt.title('Dois Gráficos Aleatórios')
plt.xlabel('Valores de eixo x')
plt.ylabel('valores do eixo y')
plt.plot(x, y1, label = 'log(x)')
plt.plot(x, y2, label = 'sen(x)/x')
plt.legend()
```

```python
x = np.arange(0.5,10, 0.001)

y1 = np.log(x)
y2 = 5 * np.sin(x) / x 

#Escolhe o modelo do gráfico
plt.style.use('seaborn')

plt.title('Dois Gráficos Aleatórios')
plt.xlabel('Valores de eixo x')
plt.ylabel('valores do eixo y')
plt.plot(x, y1, label = 'log(x)')
plt.plot(x, y2, label = 'sen(x)/x')
plt.legend()
```

```python
#Algumas formatações de gráfico 
x = np.arange(0.5,10, 0.001)

y1 = np.log(x)
y2 = 5 * np.sin(x) / x 

#Escolhe o modelo do gráfico
plt.style.use('ggplot')

#Auterar as dimensões do grafico
plt.figure(figsize=(7,5))

plt.title('Dois Gráficos Aleatórios', fontsize=16, fontweight='bold',fontstyle = 'italic', fontfamily='serif')
plt.xlabel('Valores de eixo x', fontsize=14)
plt.ylabel('valores do eixo y',fontsize=14)

#Da um espaçamento mehor
plt.tight_layout()

plt.plot(x, y1, label = 'log(x)')
plt.plot(x, y2, label = 'sen(x)/x')
plt.legend()
```
